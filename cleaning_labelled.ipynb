```python
import pandas as pd
import re
```

```python
data = pd.read_csv("Tweets.csv",index_col=0)
```

```python
clean_tweets = []

emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)

for tweet in list(data.text):
    tweet = str(tweet)
    
    #remove mentions
    tweet = re.sub(r'@[^ ]+', '', tweet)
    
    #remove https..
    tweet = re.sub(r'https?://[^ ]+', '', tweet)
    
    tweet=re.sub(r'[,.;@#?!&:$\-)(\']+', ' ', str(tweet), flags=re.IGNORECASE)
    tweet=re.sub(' +', ' ', str(tweet), flags=re.IGNORECASE)
    tweet=re.sub(r'\"', ' ', str(tweet), flags=re.IGNORECASE)
    tweet=re.sub(r'[^a-zA-Z.]', " ", str(tweet), flags=re.VERBOSE)
    tweet = emoji_pattern.sub(r'', tweet)
    tweet = " ".join(tweet.split())
    clean_tweets.append(tweet.lower())
```

```python
clean_tweets
```

```python
data.text = clean_tweets
```

```python
data.to_csv("cleaned_labelled.csv")
```
