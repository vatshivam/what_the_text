```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
URL = "https://www.revechat.com/blog/examples-of-excellent-customer-service-in-the-airline-industry/"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
job_elements = soup.find_all("p")
```

```python
all_sentences = []
for element in job_elements:
    print(element)
    element = str(element.get_text())
    element = element.strip()
#     print("text:",element)
    element=re.sub(r'[,;@#?!&$\-\']+', ' ', str(element), flags=re.IGNORECASE)
#     print("text:",element)
    element=re.sub(' +', ' ', str(element), flags=re.IGNORECASE)
#     print("text:",element)
    element=re.sub(r'\"', ' ', str(element), flags=re.IGNORECASE)
#     print("text:",element)
    element=re.sub(r'[^a-zA-Z.]', " ", str(element), flags=re.VERBOSE)
#     print("text:",element)
    element=re.sub(' +', ' ', str(element), flags=re.IGNORECASE)
    element=element.replace(',', '')
    element = element.lower()
    print(element)
    sentences = element.split(".")
    for sentence in sentences:
        if len(sentence)>1:
            all_sentences.append(sentence.strip())
```

```python
data = pd.DataFrame({'sentences':all_sentences})
```

```python
data.to_csv("scraped_data.csv")
```
